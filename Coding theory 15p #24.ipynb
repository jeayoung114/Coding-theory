```python
Found_bit_set =[]
Found_bit_set.append(0)

def search(start_bit, Found_bit_set):
    for bit in range(start_bit,64): # 0~63

        count = 0

        for target_bit in Found_bit_set: # set의 원소를 한번씩 훑기
            XOR = target_bit^bit        
            dist = 0
            for i in range(6):
                if ((1<<i) & XOR):
                    dist=dist+1

            # 3미만(0,1,2)이면 취급 안함
            if dist < 3:
                break
            else:
                count=count+1

        if count==len(Found_bit_set):
            Found_bit_set.append(bit)
            
            # 8개인 것들이 무엇이 있는지 찾아보자#
            if len(Found_bit_set) == 8:
                print("#8개인 것들...")
                for found_bit in Found_bit_set:
                    print(bin(found_bit))
                print("\n")
            ##############################
            
            # 9개인 것들이 있으면 출력하자! 하지만 없구나...
            if len(Found_bit_set) == 9:
                for found_bit in Found_bit_set:
                    print(bin(found_bit))
                return 1          
            
            if search(bit, Found_bit_set) == 1:
                return
            else:
                Found_bit_set.remove(bit)         

            
search(0,Found_bit_set)

```
